# Web Scraping
requests is mostly for getting stuff across the network, and BeautifulSoup is for looking at it.

```python
import requests
from bs4 import BeautifulSoup
```

Replace https://loc.gov with whatever site you want to scrape:

```python
page = requests.get("https://loc.gov").content
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

Now I'm going to try to find something on that page:

```python
page_title = soup.find('title').get_text()
```

```python
print(page_title)
```

I can also just print out the page code in a way that's readable:

```python
print(soup.prettify())
```
